```python
class ATMMachine:
  #static variables
  __customer_id=1
  def __init__(self):#constructor function

    print(id(self))#Represents id in memory
    self.pin=""
    self.balance=0
    self.__customer_id=ATMMachine.__customer_id
    ATMMachine.__customer_id+=1
    #self.customer_id+=1

    #self.menu()
  
  def get_customer_id(self):
    return self.__customer_id

  def menu(self):#Menu is a method of the ATMMachine class
    user_input=input("""
    Hello, how would you like to proceed?
    1. Enter 1 to create pin
    2. Enter 2 to change pin
    3. Enter 3 to check balance
    4. Enter 4 to withdraw
    """)

    if user_input=="1":
      self.create_pin()
    elif user_input=="2":
      self.change_pin()
    elif user_input=="3":
      self.check_balance()
    elif user_input=="4":
      self.withdraw()
    else:
      exit()

  def create_pin(self):#We are mentioning self here becuase it is a method of the atmmachine class
    user_pin=int(input("Enter your pin: "))
    self.pin=user_pin

    user_balance=int(input("Enter balance: "))
    self.balance=user_balance

    print("Pin created succesfully")
    #self.menu()

  def change_pin(self):
    old_pin=int(input("Enter the old pin: "))
    if old_pin==self.pin:
      new_pin=int(input("Enter the new pin: "))
      self.pin=new_pin
      print("Pin changed succesfully.")
    else:
      print("Invalid pin")
      #self.menu()

  def check_balance(self):
    user_pin=int(input("Enter your pin: "))
    if user_pin==self.pin:
      print("Your balance is: ",self.balance)
    else:
      print("Invalid pin.")
      #self.menu()

  def withdraw(self):
    User_pin=int(input("Enter your pin: "))
    if User_pin==self.pin:
      #Allow to withdraw
      amount=int(input("Enter the amount you want to withdraw: "))
      if amount> self.balance:
        print("Insufficient balance")
      elif amount<=self.balance:
        self.balance=self.balance-amount
        print("Withdraw succesfully")
    else:
      print("Invalid pin")
      #self.menu()
```
