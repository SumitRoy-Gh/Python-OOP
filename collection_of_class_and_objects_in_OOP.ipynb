```python
class person:
    def __init__(self,name,country):
        self.name=name
        self.country=country
```

```python
p1=person("Sumit","ind")
p2=person("Hero","Utopia")
p3=person("Nayak","lithuania")
```

```python
list=[p1,p2,p3]
```

```python
print(list)
```

```python
for i in list:
    print(i.name,i.country)
```
